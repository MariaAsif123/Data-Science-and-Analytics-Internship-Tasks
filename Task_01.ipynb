# **Task 01: Exploring and Visualizing a Simple Dataset**

1. Introduction & Problem Statement




```python
# The objective of this task is to explore and visualize a simple dataset
# to understand its structure, relationships between variables, and identify patterns.
# We will use the Iris dataset â€” a well-known dataset in machine learning and statistics.
# The dataset contains measurements of iris flowers (sepal length, sepal width, petal length, petal width)
# from three different species: Setosa, Versicolor, and Virginica.
```

2. Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

3. Dataset Understanding and Description

```python
# Loading dataset using pandas
iris = pd.read_csv("/Iris.csv")

# Display basic dataset information
print("Dataset Shape:", iris.shape)
print("Columns:", iris.columns.tolist())
print("\nFirst 5 Rows:\n", iris.head())

# Dataset Description:
# - SepalLengthCm: Length of the sepal (in cm)
# - SepalWidthCm: Width of the sepal (in cm)
# - PetalLengthCm: Length of the petal (in cm)
# - PetalWidthCm: Width of the petal (in cm)
# - Species: Type of iris flower (Setosa, Versicolor, Virginica)
```

4. Data Cleaning and Preparation

```python
# Check for missing values
print("\nMissing Values:\n", iris.isnull().sum())

# Check data types
print("\nData Types:\n", iris.dtypes)

# In this dataset:
# - There are no missing values.
# - All numerical columns are of float/int type and categorical is object type.
# No cleaning is required.
```

5. Exploratory Data Analysis (EDA) with Graphs

```python
# Scatter Plot: Sepal Length vs Sepal Width
sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species", palette='Set2')
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()

# Histogram: Petal Length Distribution
sns.histplot(data=iris, x="PetalLengthCm", kde=True, bins=15, color='skyblue')
plt.title("Distribution of Petal Length")
plt.show()

# Box Plot: Petal Width by Species
sns.boxplot(data=iris, x="Species", y="PetalWidthCm", palette='pastel')
plt.title("Petal Width by Species")
plt.show()

```

6. Conclusion

```python
# - The dataset contains 150 samples of iris flowers equally distributed among 3 species.
# - Sepal length and width show clear separation for Setosa, while Versicolor and Virginica overlap more.
# - Petal length distribution varies significantly across species, with Setosa having distinctly smaller petals.
# - Petal width is smallest for Setosa and largest for Virginica.
# These insights can guide further classification model development.
```
